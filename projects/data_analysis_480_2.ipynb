```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table

%matplotlib inline
```

```python
line_list = [line.rstrip('\n') for line in open('i-band.list')]
n = 100
iband = []

for i in np.arange(n):
    iband_file = np.genfromtxt(line_list[i])
    iband.append(Table(names=['time', 'imag', 'err'], data=iband_file))
```

```python
plt.scatter(iband[49]['time'], iband[49]['imag'])
plt.errorbar(iband[49]['time'], iband[49]['imag'], yerr=iband[49]['err'], fmt='none',
            ecolor='gray')
```

```python
# making plot of the folded lightcurve using a different method
# print(line_list[49], iband[49])
period = 1050

foldedtime = iband[49]['time'] / period
phase = foldedtime % 1
print(phase)

plt.scatter(phase, iband[49]['imag'])
plt.errorbar(phase, iband[49]['imag'], yerr=iband[49]['err'], fmt=' ',
            ecolor='gray')
```

```python
cepheid_test_dat = np.genfromtxt('test_OGLE-LMC-CEP-0002.dat')
cepheid_test = Table(names=['time', 'imag', 'err'], data=cepheid_test_dat)
cepheid_test
```

```python
period = 3.1191195
foldedtime_cepheid = cepheid_test['time'] / period
phase_cepheid = foldedtime_cepheid % 1

plt.figure(figsize=(10, 10))
plt.scatter(phase, iband[49]['imag'], label='mira')
plt.errorbar(phase, iband[49]['imag'], yerr=iband[49]['err'], fmt=' ',
            ecolor='gray', label='mira_err')
plt.show()
plt.figure(figsize=(10, 10))
plt.scatter(phase_cepheid, cepheid_test['imag'], label='cepheid')
plt.errorbar(phase_cepheid, cepheid_test['imag'], yerr=cepheid_test['err'],
             fmt=' ', ecolor='gray', label='cepheid_err')
plt.legend(loc=1)
plt.xlabel('Phase')
plt.ylabel('i-magnitude')
plt.title('Cepheid Variable')
plt.savefig('cepheid_variable')
plt.show()
```

```python
c_rich_MIRA = Table.read('c-rich-mira.csv')
cepheid_data = Table.read('cepheids.csv', format='ascii')
o_rich_MIRA = Table.read('o-rich-list_fixed.csv')
```

```python
c_rich_MIRA
```

```python
fig,ax = plt.subplots(1,1)
fig.set_size_inches(5,5)

mask1 = o_rich_MIRA['P_1'] < 600
mask2 = c_rich_MIRA['P_1'] < 600

ax.hist(cepheid_data['P_1'], bins=30, histtype='step', normed=True
       , label='Cepheids')
ax.hist(o_rich_MIRA['P_1'][mask1], bins=30, histtype='step', normed=True
       , label='O-rich Miras')
ax.hist(c_rich_MIRA['P_1'][mask2], bins=30, histtype='step', normed=True
       , label='C-rich Miras')

ax.legend(loc=1)
ax.set_title('Periods histogram', fontsize=15)
ax.set_xlabel('Periods', fontsize=12)
ax.set_ylabel('Normalized number of stars', fontsize=12);
# fig.savefig('period_hist_wcepheid', bbox_inches='tight');
```

```python
fig,ax = plt.subplots(1,1)
fig.set_size_inches(5,5)

ax.hist(o_rich_MIRA['P_1'], bins=30, histtype='step', normed=True
       , label='O-rich Miras', color='orange')
ax.hist(c_rich_MIRA['P_1'], bins=30, histtype='step', normed=True
       , label='C-rich Miras', color='green')

ax.legend(loc=1)
ax.set_title('Periods histogram', fontsize=15)
ax.set_xlabel('Periods', fontsize=12)
ax.set_ylabel('Normalized number of stars', fontsize=12);
# fig.savefig('periods_hist', bbox_inches='tight');
```
