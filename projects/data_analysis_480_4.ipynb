```python
import numpy as np
from astropy.table import Table
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
m33_ceph = Table.read('m33.csv')
m33_mira = Table.read('m33_mira.csv')
galactic_mira = Table.read('galactic_bulge_miras.csv')
```

```python
period1 = m33_ceph['Per']
imag1 = m33_ceph['Imag']

plt.scatter(period1, imag1, marker='.')
plt.ylim(plt.ylim()[::-1])
plt.xlabel('Period(d)')
plt.ylabel('Imag')
plt.savefig('images/m33_ceph', bbox_inches='tight')
```

```python
period2 = m33_mira['Per']
imag2 = m33_mira['<Imag>']

plt.scatter(period2, imag2, marker='.')
plt.ylim(plt.ylim()[::-1])
plt.xlabel('Period(d)')
plt.ylabel('Imag')
plt.title('M33 Mira candidates')
```

```python
period3 = galactic_mira['Per']
imax3 = galactic_mira['Imax']
imin3 = galactic_mira['Imin']
imag3 = (imax3 + imin3) / 2

plt.scatter(period3, imag3, marker='.', label='Galactic Bulge')
plt.scatter(period2, imag2, marker='.', label='M33')
plt.ylim(plt.ylim()[::-1])
plt.xlabel('Period(d)')
plt.ylabel('Imag')
plt.legend(loc='best')
plt.savefig('images/m33_galactic_miras')
# plt.xscale('log');
```
